```python
import pandas as pd
```

```python
iris = pd.read_csv('/content/Sample Dataset.csv')
iris
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(0,20)
y= x**2
plt.plot(x, y)
plt.xlabel('x-values')
plt.ylabel('x^2-values')
plt.title('Fig. 1 Line GRaph ')
plt.grid(True)
plt.show()
```

```python
iris.plot(kind='line',x='NLP', y='AI')
```

```python
iris.plot(kind='scatter', x='NLP', y='AI') 
plt.show()
```

```python
iris.plot(kind='scatter', x='NLP', y='AI',color='red',grid='True',title='Scatter Plot') 
plt.show()
```

```python
colours = {'A':'orange', 'B':'green', 'C':'blue'}
for i in range(len(iris['Machine Learning'])):
  plt.scatter(iris['NLP'][i],iris['AI'][i], color = colours[iris['Section'][i]])
plt.title('Scatter Plot')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.grid(True)
plt.show()
```

```python
iris.plot(kind='box', x='NLP', y='AI') 
plt.show()
```

```python
a= iris['Section'].value_counts()
grades = a.index
grades
```

```python
a= iris['Section'].value_counts()
grades = a.index
count = a.values
plt.bar(grades,count,color = 'blue')
plt.xlabel('grades')
plt.ylabel('number of students')
plt.show()
```

```python
length_width = iris[['Machine Learning','NLP','AI','Section']] #excluding species column
length_width.boxplot()
plt.xlabel('Measurements')
plt.ylabel('values')
plt.title("Dataset analysis")
plt.show()
```

```python
iris.plot(kind='hist',x='NLP', y='AI')
```

```python
iris.plot.hist( layout=(2,2), figsize=(10, 10), bins=20)
```

```python
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10, 10), bins=20)
```

```python
a= iris['Section'].value_counts()
species = a.index
count = a.values
colors= ['lightblue','darkgreen','yellow']
explode = (0,0.6,0)
plt.pie(count, labels=species,shadow=False,
        colors=colors,explode = explode, autopct='%1.1f%%')
plt.xlabel('species')
plt.axis('equal')
plt.show()
```
